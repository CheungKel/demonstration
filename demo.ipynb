```python
stock_data = pd.read_csv("sorted_stock.csv")
stock_data
```

```python
import matplotlib.pyplot as plt
```

```python
year_zero4 = stock_data[stock_data.Year == 2004]
year_zero4
```

```python
mean_list = [0]
def mean(x):
    for i in range(1,x+1):
        mean = year_zero4["Close"].iloc[:i+1].mean()
        mean_list.append(mean)
    return mean_list
mean(93)
year_zero4["Mean"] = mean_list

median_list = [0]
def median(x):
    for i in range(1,+x+1):
        median = year_zero4["Close"].iloc[:i+1].median()
        median_list.append(median)
    return median_list
median(93)
year_zero4["Median"] = median_list

up_down_list = [0]
def up_down(x):
    for i in range(1,x+1):
        up_down = year_zero4["Close"].iloc[i]-year_zero4["Close"].iloc[i-1]
        up_down_list.append(up_down)
    return up_down_list
up_down(93)
year_zero4["Up/Down"] = up_down_list

change_percentage_list = [0]
def change_percentage(x):
    for i in range(1,x+1):
        change_percentage = up_down_list[i]/year_zero4["Close"].iloc[i-1]*100
        change_percentage_list.append(change_percentage)
    return change_percentage_list
change_percentage(93)
year_zero4["Change Percentage"] = change_percentage_list
year_zero4
```

```python
plt.figure(figsize = (40,20))
mean = year_zero4["Mean"]
median = year_zero4["Median"]

plt.plot(year_zero4.Date[::3], mean[::3], label = "Mean")
plt.plot(year_zero4.Date[::3], median[::3], label = "Median")

plt.xlabel("Date")
plt.ylabel("Stock Index")
plt.legend()
```

```python
plt.figure(figsize = (40,20))
up_down = year_zero4["Up/Down"]

plt.plot(year_zero4.Date[::3], up_down[::3], label = "Up/Down")
plt.xlabel("Date")
plt.ylabel("Up/Down")
plt.legend()
```

```python
plt.figure(figsize = (40,20))
change_percentage = year_zero4["Change Percentage"]

plt.plot(year_zero4.Date[::3], change_percentage[::3], label = "Change Percentage")
plt.xlabel("Date")
plt.ylabel("(%)")
plt.legend()
```
